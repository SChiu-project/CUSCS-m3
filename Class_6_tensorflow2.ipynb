Repeat the dataset for each epoch

```python
z = y.repeat(count=2)
z
```

```python
for row in z:
    print('x:', row[0].numpy(),
          'y:', row[1].numpy())
```

```python
z2 = z.batch(3)
for row in z2:
  print('x:', row[0].numpy(),
      'y:', row[1].numpy())
```

```python
z3 = y.repeat(count=2).batch(3)
for row in z3:
    print('x:', row[0].numpy(),
          'y:', row[1].numpy())
```

```python
z4 = y.batch(3).repeat(count=2)
for row in z4:
    print('x:', row[0].numpy(),
          'y:', row[1].numpy())
```

```python
z5 = y.repeat(count=2).shuffle(4).batch(3)
for row in z5:
    print('x:', row[0].numpy(),
          'y:', row[1].numpy())
```

linear regression

```python
# Keras is the second level
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
layer0 = Dense(units=1, input_shape=[1])
model = Sequential([layer0])
model.compile(optimizer='sgd',loss='mean_squared_error')

X = np.array([-1.,0., 1., 2., 3., 4.])
y=np.array([-3., -1., 1., 3., 5. , 7.])
```

```python
plt.plot(X,y)
```

```python
model.fit(x=X, y=y, epochs=500)
```

```python
layer0.get_weights()
```

```python
model.predict([0.5])
```

```python
model.predict([10])
```

```python
class MyModel(tf.keras.Model):
  def __init__(self):
    super(MyModel, self).__init__()
    self.w = tf.Variable(0., name='weight')
    self.b = tf.Variable(0., name='basis')
  def call(self,x): # add a function on top of keras model
    return self.w*x + self.b
```

```python
model = MyModel()
model.build(input_shape=(None, 1))
model.summary()
```
